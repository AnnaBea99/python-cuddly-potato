## Tensors

"A tensor is a generatization of vectors and matrices to potentialli higher dimensions. Internally, TensorFlow reporesents tensors as n-dimensional arrays of base datatypes". (http://tensorflow.org/guide/tensor)

It should't surprise you that tensors are a fundamental aspect of TensorFlow. They are the mais objects that are passed around and manipulate throughout the program. Each tensor represents a partialy defined computation tha will eventually produce a value. TensorFlow programs work by building a graph of Tensor objects that details how tensors are related. Running different parts of the graph allow results to be generated.

Each tensor has a data type and a shape.

**Data Types Include:** float32, int32, string and others.

**Shape:** Represents the dimension of data.
